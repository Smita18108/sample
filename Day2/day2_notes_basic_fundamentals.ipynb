**Data types**

```python
# In Python, there is no separate char data type like in some other languages (java)

num = 12
print(type(num))

num2 = 1.2
print(type(num2))


name = 'smita'
print(type(name))

value = True
print(type(value))

Division = "A"
print(type(Division))

print(num,num2,name,value,Division)
```

**Arithmetic Operators**

```python
# split() method used to split a string into a list of substrings.
x, y = input("Enter any two numbers: ").split(', ')

# Type Casting
x = int(x)
y = int(y)

print("Sum of two numbers is : ", x + y)
print("Sub of two numbers is : ", x - y)
print("Multi of two numbers is : ", x * y)
print("Div of two numbers is : ", x / y)
print("Modulus of two numbers is : ", x % y)
```

**Assignment operators**

```python
x = 10
print(x)
x += 5
print(x)
x -= 3
print(x)
x *= 3
print(x)
x /= 4
print(x)
x %= 2
print(x)
```

**Comparison operators**

```python

x = 10
y = 10
print("x = ", x, "Y = ", y)
print("x is equal to y :", x == y)
print("x is not equal to y :", x != y)
print("x is less than y :", x < y)
print("x is greater than y :", x > y)
print("x is less than equal to y :", x <= y)
print("x is greater than equal to y :", x >= y)
```

**Logical operators**

```python
a = 10
b= 5
# NOT AND
if not (a > b and a!=b):
    print("Hello")
else:
    print("Bye bye")

# NOT OR
if not (a < b or a==b):
    print("Hello")
else:
    print("Bye bye")

# OR
if (a > b or a==b):
    print("Hello")
else:
    print("Bye bye")

# AND
if (a < b and a==b):
    print("Hello")
else:
    print("Bye bye")
```

**Conditional statements**

```python

a = 5
b = 5
if a > b:
    print("a is greater than b")
elif a < b:
    print("a is less than b")
else:
    print("a is equal to b")
```
