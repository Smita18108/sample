**creating a list**

```python
list = [10,20,15,40,5]
print("list : ", list)
```

**Methods :**

**1. append()**

```python

list.append(50)
print(list)
```

```python
new_list = ["toyoto", "honda"]
```

```python
new_list.append(list)
print(new_list)
```

**2. extend()**

```python
num = [1,2,3,4,5]
```

```python
name = ["kanha", "radhe", "shyam"]
num.extend(name)
print(num)
```

**3. insert()**

```python
num1 = [1,2,3,4,5]
num1.insert(3, 10) # must write 2 arguments while using insert(index, value) method.
print(num1)
```

**4. remove()**

```python
name = ["abc", "def", "ghi", "abc"]
name.remove("abc")
print(name)
```

**5. pop()**

```python
num2 = [1,2,3,4,5]
popped_value = num2.pop()
print("popped_value :", popped_value)
print(num2)
```

```python
name = ["abc", "def", "ghi", "abc"]
popped_value = name.pop()
print("popped_value :", popped_value)
print(name)
```

**6. clear()**

```python
print(name)
name.clear()
print(name)
```

**7. index()**

```python
num3 = [1,2,3,4,5]
print(num3.index(4))
```

**8. count()**

```python
num4 = [4,2,3,4,5,1,4,3,4,5, 2, 4, 6]
print("count of 4 : ", num4.count(4))
```

**9.reverse()**

```python
str = ["kanha", "shyam", "radhe"]
str.reverse()
print(str)
```

**10. sort()**

```python
str = ["kanha", "shyam", "radhe"]
str.sort()
print(str)
```

```python
num5 = [3,5,3,4,2,1]
print(num5)
num5.sort()   # default ascending order
print(num5)
```

```python
num5.sort(reverse=True)  # to make it in discending order write reverse=True.
print(num5)
```

**11. copy()**

```python
num_list = [1,2,3,4]
copy_list = num_list.copy()
print("Coppied list : ", copy_list)
```

**12. len()**

```python
len(num_list)
```

**13. sum()**

```python
name2 = ["kanha", "radhe", "shyam"] # This method only works on integers not strings.
sum(name2)
print(sum(name2))
```

```python
num = [1,2,3,4,5]
print(sum(num))
```

**14. min() or max()**

```python
print(max(num))
print(min(num))
```
