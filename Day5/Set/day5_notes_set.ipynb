**Set**

Set - unordered, unindexed, mutable collection of unique elements (non duplicate elements)

characters :
mutable,

operations :
add, update, remove, discard, pop, clear, union, intersection, difference, symmetric_difference, issubset, issuperset, len

```python
my_set = {1, 2, 3, 4, 5}
print(my_set)
```

**Functions in set**

1. add()

```python
my_set = {1, 2, 3, 4, 5}
my_set.add(6)
print(my_set)
```

2. update() - converts the data structure into set as well as update the set.

```python
my_set = {1, 2, 3, 4, 5}
my_set.update([6, 7, 8])
print(my_set)
```

**3. remove()**

```python
my_set = {1, 2, 3, 4, 5}
my_set.remove(3)
print(my_set)
```

**4. discard()** - wont show any error as happens in remove() after giving invalid value.

```python
my_set = {1, 2, 3, 4, 5}
my_set.discard(6)
print(my_set)
```
