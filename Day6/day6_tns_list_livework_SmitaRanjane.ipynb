**1. Write a python function that takes a list of integers and returns a new list with all the even numbers doubled**

```python
def list_even_num(new_list):
  print(new_list)
  new_list = [int(x) for x in new_list]
  new_list = [x * 2 for x in new_list if x % 2 == 0]
  return new_list

new_list = input("Enter the list of integers: ").split(',')
print(list_even_num(new_list))
```
